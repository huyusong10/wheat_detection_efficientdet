```python
from glob import glob
# name_list = glob('*.txt')
name_list = ['baseline.txt', '>20.txt', '<20.txt', '%2.txt', '%4.txt']
```

```python

ls = []

for i in name_list:
    with open(i) as f:
        l = np.array([x.split() for x in f.read().splitlines()])   
        ls.append(l)  
```

```python
figsize = (8, 6)
index = list(range(100))
p = [x[:, 15].astype(np.float32) for x in ls]
r = [x[:, 17].astype(np.float32) for x in ls]
f = [x[:, 19].astype(np.float32) for x in ls]
mAP = [x[:, 21].astype(np.float32) for x in ls]
```

```python
plt.figure(figsize=figsize)
plt.title("baseline", fontsize=20)
plt.xlabel('epoch', fontsize=20)
plt.ylim(0, 1)
ax=plt.gca()
ax.yaxis.set_major_locator(MultipleLocator(0.1))
plt.plot(index, p[0], label='precision', color='blue')
plt.plot(index, r[0], label='recall', color='green')
plt.plot(index, mAP[0], label='mAP:0.5', color='red')
plt.legend(loc = 'best', fontsize=20)
```
